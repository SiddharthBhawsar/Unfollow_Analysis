```python
# Calculate the sum of tone_pos and tone_neg
total_tone = df['tone_pos'].sum() + df['tone_neg'].sum()

# Calculate the percentage of tone_pos and tone_neg
percent_pos = (df['tone_pos'].sum() / total_tone) * 100
percent_neg = (df['tone_neg'].sum() / total_tone) * 100

# Print the results
print(f"Sum of tone_pos and tone_neg: {total_tone}")
print(f"Percentage of tone_pos: {percent_pos:.2f}%")
print(f"Percentage of tone_neg: {percent_neg:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Calculate the sum of each emotion
total_emo_pos = df['emo_pos'].sum()
total_emo_neg = df['emo_neg'].sum()
total_emo_anx = df['emo_anx'].sum()
total_emo_anger = df['emo_anger'].sum()
total_emo_sad = df['emo_sad'].sum()

# Calculate the percentage of each emotion
percent_emo_pos = (total_emo_pos / (total_emo_pos + total_emo_neg + total_emo_anx + total_emo_anger + total_emo_sad)) * 100
percent_emo_neg = (total_emo_neg / (total_emo_pos + total_emo_neg + total_emo_anx + total_emo_anger + total_emo_sad)) * 100
percent_emo_anx = (total_emo_anx / (total_emo_pos + total_emo_neg + total_emo_anx + total_emo_anger + total_emo_sad)) * 100
percent_emo_anger = (total_emo_anger / (total_emo_pos + total_emo_neg + total_emo_anx + total_emo_anger + total_emo_sad)) * 100
percent_emo_sad = (total_emo_sad / (total_emo_pos + total_emo_neg + total_emo_anx + total_emo_anger + total_emo_sad)) * 100

# Plot the results as a pie chart
labels = ['Positive', 'Negative', 'Anxiety', 'Anger', 'Sadness']
sizes = [percent_emo_pos, percent_emo_neg, percent_emo_anx, percent_emo_anger, percent_emo_sad]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Emotion Percentage')
plt.show()
```
